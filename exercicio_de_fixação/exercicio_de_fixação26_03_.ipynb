```python
#1 Leia a nota de um aluno. Classifique usando elif:
#nota ≥ 9 → "A — Excelente", ≥ 7 → "B — Aprovado", ≥ 5 → "C — Recuperação", abaixo → "D — Reprovado".
nota = float(input("Digite sua Nota:"))
if nota >= 9:
    print("A - Excelente")
elif nota >= 7:
    print("B - Aprovado")
elif nota >= 5:
    print("C - Recuperação")
else:
    print("D - Reprovado")
```

```python
#2 Leia um número de 1 a 7 e exiba o nome
#do dia correspondente (1 = Segunda, ..., 7 = Domingo). Para qualquer outro número, exiba "Número inválido".
dia = int(input("Digite um número de 1 a 7:"))
match dia:
    case 1: print("Segunda Feira")
    case 2: print("Terça Feira")
    case 3: print("Quarta Feira")
    case 4: print("Quinta Feira")
    case 5: print("Sexta Feira")
    case 6: print("Sábado")
    case 7: print("Domingo")
    case _ : print(" Dia da Semana não Existe !")
```

```python
#3 Leia uma nota. Valide se está entre 0 e 10 usando encadeamento de comparações
#(0 <= nota <= 10). Se inválida, exiba "Nota inválida!". Se válida, exiba "Nota registrada: [nota]".
nota = float(input("Digite uma nota: "))
if 0 <= nota <= 10:
    print(f"Nota registrada: {nota}")
    print("Nota válida")
else:
    print("Nota inválida!")
```

```python
#4 Leia a idade. Classifique: menor de 12 → "Criança",
#12 a 17 → "Adolescente", 18 a 59 → "Adulto", 60 ou mais → "Idoso". Valide que a idade não é negativa.
idade = int(input("Digite sua idade: "))
if idade < 0 or idade > 120:
    print("Idade inválida!")
elif idade >= 1 and idade < 12:
    print("Criança")
elif idade >= 12 and idade <18:
    print("Adolescente")
elif idade >= 18 and idade <60:
    print("Adulto")
elif idade >= 60:
    print("Idoso")


```

```python
#6 Leia a temperatura. Classifique usando elif: abaixo de 0 → "Congelante", 0 a 15 → "Frio", 16 a 25 → "Agradável", 26 a 35 → "Quente", acima de 35 → "Muito quente".
temp = float(input("Digite a Temperatura: "))
if temp < 0:
    print("Temperatura Congelante!")
elif temp >= 1 and temp < 15:
    print("Frio")
elif temp >= 16 and temp <25:
    print("Agradável")
elif temp >= 26 and temp <35:
    print("Quente")
elif temp >= 35:
    print("Muito Quente")
```

```python
#7 Leia dois números e um operador (+, -, *, /). Use elif para realizar a operação correta.
#Valide a divisão por zero. Para operador inválido, exiba "Operador desconhecido".
numA = float(input("Número 1:"))
operador = input("Operador:")
numB = float(input("Número 2:"))
if operador == "+": print(f"{numA} {operador} {numB} = {eval(str(numA)+operador+str(numB))}")
elif operador == "-": print(f"{numA} {operador} {numB} = {eval(str(numA)+operador+str(numB))}")
elif operador == "*": print(f"{numA} {operador} {numB} = {eval(str(numA)+operador+str(numB))}")
elif operador == "/": print(f"{numA} {operador} {numB} = {eval(str(numA)+operador+str(numB))}")
else:
    print("Operação desconhecido")

```

```python
#9 Leia três lados. Valide primeiro se todos são maiores que zero.
#Em seguida, verifique se formam um triângulo (cada lado menor que a soma dos outros dois). Exiba o resultado.
ladoA = float(input("Digite tamanho do lado A:"))
ladoB = float(input("Digite tamanho do lado B:"))
ladoC = float(input("Digite tamanho do lado C:"))
if ladoA > 0 and ladoB > 0 and ladoC > 0:
    if (ladoA + ladoB) > ladoC and (ladoA + ladoC) > ladoB and (ladoC + ladoC) > ladoA:
        print("Forma um triângulo")
    else:
        print("Não forma um triângulo")
```
