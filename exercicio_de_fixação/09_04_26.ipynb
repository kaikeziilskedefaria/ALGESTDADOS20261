```python
# 1 Usando while, exiba os números de 1 até 10, um por linha.
# Use uma variável de controle iniciada em 1 e incremente a cada iteração.
contador = 1
while contador <= 10 :
    print(contador)
    contador += 1
```

```python
# 2 Defina uma senha correta no código. Usando while, peça a senha ao usuário repetidamente até que ele acerte.
# A cada tentativa errada, exiba 'Senha incorreta, tente novamente.'.
# Ao acertar, exiba 'Acesso liberado!' e quantas tentativas foram necessárias.
contador = 0
senha_correta = "12345"
while True :
    senha = input("Digite a senha: ")
    contador += 1
    if senha == senha_correta :
        print("acesso liberado !")
        print(f"Tentativas necessárias: {contador}")
        break
    else :
        print("Senha incorreta, tente novamente.")
    contador += 1
```

```python
# 3 Defina um número secreto entre 1 e 100 no código. Usando while, leia tentativas do usuário até acertar.
# A cada erro, diga se o número secreto é maior ou menor. Ao final, exiba quantas tentativas foram usadas.
contador = 0
número_secreto = "8"
while contador <= 100:
    while True:
        tentativa = int(input("Digite um número: "))
        contador += 1
        break
    if tentativa == int(número_secreto) :
        print("acertou")
        print(f"Tentativas necessárias: {contador}")
        break
    else:
        if tentativa > int(número_secreto) :
            print("O número secreto é menor")
            contador += 1
        elif tentativa < int(número_secreto) :
            print("O número secreto é maior")
```
