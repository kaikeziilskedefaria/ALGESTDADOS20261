```python
#1 Peça um número inteiro ao usuário. Se o número
#for maior que zero, exiba a mensagem "O número é positivo!".
num = int(input("digite um número inteiro: "))
if num > 0:
    print("O número é positivo!")
```

```python
#2 Leia a idade do usuário. Se for maior ou igual a 18, exiba
#"Maior de idade", caso contrário, exiba "Menor de idade".
idade = int(input(" Digite sua idade: "))
if idade >= 18:
    print("Maior de idade")
else:
    print("Menor de idade")
```

```python
#3 Solicite um número inteiro. Verifique se ele é par
#(divisível por 2) e exiba uma mensagem informando o resultado.
num = int(input(" Digite um número inteiro :"))
if num % 2 == 0:
    print("O número é par")
else:
    print("O número é ímpar")
```

```python
#4 Defina uma senha fixa no código (ex: "python123"). Peça a senha ao usuário.
#Se bater, exiba "Acesso permitido", senão "Acesso negado".
senha_fixa = "python123"
senha = input("Digite a senha: ")
if senha == senha_fixa:
    print("Acesso permitido")
else:
    print("Acesso negado")
```

```python
#5 Leia a temperatura em graus Celsius. Se for maior que 30, exiba "Está
#quente hoje!". Caso contrário, exiba "Temperatura agradável."
temperatura = float(input("Digite a temperatura em graus Celsius(C°): "))
if temperatura > 30:
    print("Está quente hoje!")
else:
    print("Temperatura agradável.")
```

```python
#6 Defina um saldo fixo (ex: 100.0). Peça ao usuário um
#valor para saque. Se o valor for menor ou igual ao saldo, exiba
# "Saque realizado!", caso contrário, "Saldo insuficiente."
saldo = 1000
saque = float(input("Digite o valor do saque: "))
if saque <= saldo:
    print("Saque realizado!")
else:
    print("Saldo insuficiente")
```

```python
#7 Leia a nota de um aluno (0 a 10). Se a nota for maior ou igual
#a 5, exiba "Aprovado". Senão, exiba "Reprovado".
nota = float(input("Digite a nota do aluno:"))
if nota >= 5:
    print("Aprovado")
else:
    print("Reprovado")
```

```python
#8 Peça um número ao usuário.
#Exiba se o número é zero ou diferente de zero usando if/else.
num = int(input("Digite um número:"))
if num == 0:
    print("O número é zero")
else:
    print("O número é diferente de zero")
```

```python
#9 Peça um número ao usuário. Sem usar == 0, use o próprio valor
# como condição no if para exibir "O número
#é falsy (zero)" ou "O número é truthy (diferente de zero)".
num = bool(int(input("Digite um número:")))
if num:
    print("O número é truthy (diferente de zero)")
else:
    print("O número é falsy (zero)")

```

```python
#10 Peça ao usuário para digitar a cor do semáforo
#("verde" ou outra cor). Se for
#"verde", exiba "Pode avançar!", caso contrário, "Pare!".
cor = input("Digite a cor do semáforo: ")
if cor == "verde":
    print("Pode avançar!")
elif cor == "amarelo":
    print("Atenção!")
else:
    print("Pare!")
```

```python
#11 Leia o nome do usuário. Se o nome for uma string vazia (falsy),
#exiba "Nenhum nome digitado.",
#caso contrário, exiba "Olá, [nome]!" usando f-string.
nome = input(" Digite seu nome: ")
if not nome:
    print("Nenhum nome digitado.")
else:
    print(f"Olá, {nome}!")
```

```python
#12 Leia dois números inteiros. Exiba qual deles é o maior.
#Se forem iguais, exiba "Os números são iguais."
num1 = int(input("Digite o primeiro número: "))
num2 = int(input("Digite o segundo número: "))
if num1 > num2:
    print(f"O primeiro número é maior: {num1}")
elif num1 < num2:
    print(f"O segundo número é maior: {num2}")
else:
    print(f" Os números são iguais!")
```

```python
#13 Pergunte ao usuário "Está chovendo? (sim/nao)". Se a resposta
# for "sim", exiba
# "Leve o guarda-chuva!", senão, "Não precisa de guarda-chuva."
resposta = input("Está chovendo? (sim/nao): ")
if resposta == "sim":
    print("Leve o guarda-chuva!")
else:
    print("Não precisa de guarda-chuva!")
```

```python
#14 Pergunte se hoje é o aniversário do usuário
#(resposta: "sim" ou "nao"). Se sim, exiba "Parabéns!
#Você tem 20% de desconto." Senão, "Sem desconto hoje."
resposta = input("Hoje é o seu aniversário? (sim/nao): ")
if resposta == "sim":
    print("Parabéns, você tem 20% de desconto!")
else:
    print("Sem desconto hoje!")

```
